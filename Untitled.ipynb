```python
def num_after_point(x):
    s = str(x)
    if not '.' in s:
        return 0
    return len(s) - s.index('.') - 1


decimal_places = 100
beta_step_size = 0.01
beta_lower = 0.01
beta_upper= 5
beta_digits_after_decimal = num_after_point(beta_step_size)
# print(beta_digits_after_decimal)
k_limit = 10
Decimal_expansion = 5
initial_x = 1
alpha_root = -1
```
